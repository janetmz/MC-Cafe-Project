```python
#Montgomery College Cafe
def print_menu():
    x = 1
    print (f"\t\t\t\t\tCOLLEGE CAFE MENU")
# Print the names of the columns.
    print (f"COFFEES\tPrices\t\tTEAS\t\tPrices\t\tPASTRIES\t\tPrices")
# Iterate through the dictionary
    for x in range(1,7):
        print (f"""{coffee_menu.get(x):<15}\t${coffee_prices.get(x):.2f} \t\t{tea_menu.get(x)} \t${tea_prices.get(x):.2f}\t\t {pastry_menu.get(x)}\t\t${pastry_prices.get(x):.2f}""")

#************************************************************************************
def check_inventory(i):
    y = 0
    if i in coffee_menu.values():
        y =10
    elif i in tea_menu.values():
        y = 10
    elif i in pastry_menu.values():
        y =  10
    else:
        y = 0
    return y
#***********************************************************************************
def build_order (item, how_many):
    #    value = item
    if item in coffee_menu.values():
        for key, value in coffee_menu.items():
            if item == value:
                key_id = key
                price = float(coffee_prices[key_id])
                cost = price* how_many
                break
    elif item in tea_menu.values():
        for key, value in tea_menu.items():
            if item == value:
                key_id = key
                price = float(tea_prices[key_id])
                cost = price* how_many
                break          
    elif item in pastry_menu.values():
        for key, value in pastry_menu.items():
            if item == value:
                key_id = key
                price = float(pastry_prices[key_id])
                cost = price* how_many
                break
    tally.append (item)
    tally.append (how_many)
    tally.append (price)
    tally.append (cost)
    
#*****************************************************************************************
def prepare_receipt (s):
    it = tally[s]                    # the item
    hm = tally[s+1]             # the number ordered
    Cost = float(tally[s+2])  # the cost of 1 item
    total = float(tally[s+3])   # the total cost for this item (price x # ordered)
    print(f"{it}\t\t{hm} @${Cost:.2f} each\t\t${total:.2f}")
    return total
#*******************************************************************************************   
        
#set up the menu and prices
coffee_menu = {1:"Expresso",2:"Americano" ,3:"Cafe Latte" ,4:"Mocha", 5:"Cappucino" , 6:"Machiato"}
coffee_prices = {1:1.50, 2: 2.00 ,3: 2.25 , 4: 2.25 , 5: 2.50 , 6: 2.50}
tea_menu = {1:"White Tea" ,2:"Green Tea" ,3:"Oolong Tea" ,4:"Black Tea" ,5:"Earl Grey" , 6:"Matcha"}
tea_prices = {1: 1.50, 2: 2.00, 3: 2.25, 4: 2.25, 5: 2.50, 6: 2.50}
pastry_menu = {1:"Croissant" ,2:"Apple Turnover",3:"Banana Bread", 4:"Pumpkin Bread" ,5:"Cinnamon Roll" ,6:"Sticky Bun"}
pastry_prices = {1:5.50,2: 4.50 ,3: 4.25 ,4: 3.75 ,5: 4.25 ,6: 3.75}

#initialize list that keeps track of each order
tally = ["x"]  #I put "x" in the first position to avoid the problem of a counter starting at 0

#Greet the customer and ask if they wish to see the menu
customer = (input("Welcome to MC Cafe!  May I please have your name\n"))
menu_question = (input (f"Hello {customer}. Would you like to see the menu (Yes or No)\t\v"))
menu_question = menu_question.upper()

if menu_question == "Yes" or menu_question == "Y":
    print_menu()

#get the customer's order
number_ordered = 0   #number of unique items ordered
y = 0
done = False
while not done:
    item = (input(f" What would you like to order\t"))
    item = item.title()
    y = check_inventory(item)
    if y > 0:   #continue processing. If y = 0 the user will be asked if they misspelled the item and will be allowed to reenter an item
        
# I have not figured out how to capture the error if the user enters a letter instead of a number for the next question
        how_many = int(input(f"How many\t"))
# Problem here if letter entered; program terminates

        number_ordered +=1
        build_order(item,how_many)   #stores the iteme, number ordered, cost of 1 items, and total cost for items * how many
        
# error checking in case the used does not enter a Y or N for the next question
        go_on = False
        while not go_on:
            more_items = (input("Do you want to order more times(Y or N)?"))
            more_items = more_items.upper()
            if more_items == "Y" or more_items == "N":
                go_on = True
            else:
                print ("Please enter Y or N")
#end of while not go_on

        if more_items == "N":
            done = True
    else:      # user did not enter a valid item name from the menu
        print ('That item is not in the inventory. Did you misspell it?')

#***********************************************************************************
#get ready to prepare the receipt. Data is being pulled from the tally list
start = 1               #list index
x = 0                    # number of orders count
total_this_item = 0      #cost of an item X number ordered
running_balance = 0
final_total = 0
z = "*"
print (z*52)
print ("\t\t\t\tReceipt")
print ("Item\t\t\tQuantity\t\t\tTotal Cost")
while x < number_ordered:
    total_this_item = prepare_receipt(start)   #returns the total cost of the item
    x +=1
    start +=4
    running_balance = running_balance + total_this_item
#************************************************************************************
#Prepare the bill
print (z*52)
print (f"Your subtotal is:\t\t\t ${running_balance:.2f}")    
tax = running_balance * 0.06
final_total = running_balance + tax
print (f"The tax is \t\t\t\t${tax:.2f}")
print (f"With tax, your bill comes to \t ${final_total:.2f}\n")
print ("Thank you for visiting MC Cafe!")
```
